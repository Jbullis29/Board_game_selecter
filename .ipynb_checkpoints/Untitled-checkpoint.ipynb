```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from splinter import Browser
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pymongo
import datetime
import requests
import time 
```

```python
# Create my connection to Pymongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
#drop my database 
client.drop_database('board_games')
```

```python
db = client.board_games
# create collection (A.K.A Table) in my database
collection = db.games
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
bgg_url ='https://boardgamegeek.com/browse/boardgame'
browser.visit(bgg_url)
```

```python
game_dict={'name':[], 'years':[], 'number_of_players':[], 'play_time':[], 'age':[], 'complexity_rating':[], 'description':[]}
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
for i in range(1,101):
    results = soup.find('div', id=f'results_objectname{i}')
    name = results.find('a').text
    try:
        game_dict['name'].append(name)
        browser.click_link_by_partial_text(name)
        browser.back()
    except:
        print(f'Missing Game: {name}')
    time.sleep(1)
```
